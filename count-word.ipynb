```python
import string
import csv
import itertools
import numpy as np
import collections
con = []

with open("new_conserv.txt","r", encoding = "utf-8") as f4:
    while True:
        x = f4.readline()
        if not x: break 
        
        y= x[:-1].split(' ') 
        con.extend(y)
        #print(con)
        '''
        for i in y:
            print(i)
            con.append(i)
            print(con)
            break'''
            
     
    f4.close()
conserv_cnt=len(con)
```

```python
conserv_distribution=collections.Counter(con)

for key, item in conserv_distribution.items():
    conserv_distribution[key]=item/conserv_cnt
conserv_distribution
```

```python
lib = []

with open("new_liberal.txt","r", encoding = "utf-8") as f4:
    while True:
        x = f4.readline()
        if not x: break 
        
        y= x[:-1].split(' ') 
        lib.extend(y)
        #print(con)
        '''
        for i in y:
            print(i)
            con.append(i)
            print(con)
            break'''
            
     
    f4.close()
liberal_cnt=len(lib)
```

```python
liberal_distribution=collections.Counter(lib)
```

```python

for key, item in liberal_distribution.items():
    liberal_distribution[key]=item/liberal_cnt
liberal_distribution
```

```python
cnt=0
for key, item in liberal_distribution.items():
    if key in conserv_distribution:
        diff=abs(conserv_distribution[key]-liberal_distribution[key])
        #print(diff)
        if diff<0.0000001:
            cnt+=1
            lib.remove(key)
            con.remove(key)
            print(cnt)
```
