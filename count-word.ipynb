```python
import string
import csv
import itertools
import numpy as np
import collections

nonsense=[]
with open("nonsense.txt","r",encoding = "utf-8") as f:
    while True:
        x=f.readline()
        if not x: break
        nonsense.append(x[:-1])
```

```python
for word in nonsense:
    print(word)
```

```python
con = []
org_conserv=[]
with open("new_conserv.txt","r", encoding = "utf-8") as f4:
    while True:
        x = f4.readline()
        if not x: break 
        
        y= x[:-1].split(' ') 
        
        #print(nonsense)
        
        for word in nonsense:
            if word in y:
                y.remove(word)
            
        con.extend(y)
        org_conserv.append(y)
        #print(con)
        '''
        for i in y:
            print(i)
            con.append(i)
            print(con)
            break'''
            
     
    f4.close()
conserv_cnt=len(con)
```

```python
org_conserv[0]
```

```python
conserv_distribution=collections.Counter(con)

for key, item in conserv_distribution.items():
    conserv_distribution[key]=item/conserv_cnt
conserv_distribution
```

```python
lib = []
org_liberal=[]
with open("new_liberal.txt","r", encoding = "utf-8") as f4:
    while True:
        x = f4.readline()
        if not x: break 
        
        y= x[:-1].split(' ')
        for word in nonsense:
            if word in y:
                y.remove(word)
        lib.extend(y)
        org_liberal.append(y)
        #print(con)
        '''
        for i in y:
            print(i)
            con.append(i)
            print(con)
            break'''
            
     
    f4.close()
liberal_cnt=len(lib)
```

```python
liberal_distribution=collections.Counter(lib)
```

```python

for key, item in liberal_distribution.items():
    liberal_distribution[key]=item/liberal_cnt
liberal_distribution
```

```python
cnt=0
for key, item in liberal_distribution.items():
    if key in conserv_distribution:
        diff=abs(conserv_distribution[key]-liberal_distribution[key])
        #print(diff)
        if diff<0.0000001:
            cnt+=1
            for i in range(len(org_conserv)):
                if key in org_conserv[i]:
                    org_conserv[i].remove(key)
            for i in range(len(org_liberal)):
                if key in org_liberal[i]:
                    org_liberal[i].remove(key)
            
            #print(cnt)
```
